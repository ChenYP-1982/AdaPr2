## Crisp-DM

  O CRISP-DM (Cross Industry Standard Process for Data Mining) é um modelo amplamente utilizado para a condução de projetos de mineração de dados. Ele fornece uma estrutura abrangente para orientar os profissionais de dados em todas as etapas do processo, desde a compreensão do negócio até a implementação e monitoramento dos resultados. O CRISP-DM é composto por seis fases principais:

1. Compreensão do Negócio (Business Understanding): Nesta fase inicial, os objetivos e requisitos do negócio são definidos.Os principais elementos incluem identificar os objetivos do projeto, determinar as metas para a mineração de dados, definir critérios de sucesso e avaliar a situação do negócio.

2. Compreensão dos Dados (Data Understanding):Esta fase envolve a coleta inicial de dados e a familiarização com eles.As atividades incluem a coleta de dados, a exploração inicial dos dados para entender sua natureza e a qualidade dos dados, identificar problemas de qualidade e familiarizar-se com a estrutura dos dados.

3. Preparação dos Dados (Data Preparation):Aqui, os dados são preparados para a modelagem.As etapas incluem a seleção de dados relevantes, limpeza de dados (tratamento de valores ausentes, duplicados, outliers), transformação de dados (normalização, padronização) e integração de dados de várias fontes.

4. Modelagem (Modeling):Nesta fase, os modelos são construídos para prever ou identificar padrões nos dados.Isso envolve a seleção de técnicas de modelagem apropriadas, a criação de modelos usando ferramentas de modelagem, a avaliação dos modelos em relação aos objetivos do negócio e a seleção do modelo final.

5. Avaliação (Evaluation):Os modelos construídos são avaliados criticamente para garantir que atendam aos objetivos do projeto.Isso inclui a avaliação da eficácia do modelo, validação cruzada, ajuste de parâmetros e revisão do processo de modelagem.

6. Implantação (Deployment):Finalmente, os resultados do projeto são implantados na organização.Isso pode incluir a implementação de modelos em sistemas de produção, a criação de relatórios e visualizações para comunicação dos resultados e o monitoramento contínuo do desempenho do modelo.

7. Feedback (Feedback):Após a implantação dos resultados, é fundamental obter feedback sobre como esses resultados estão impactando o negócio.Nesta fase, os resultados são monitorados em relação aos objetivos de negócios definidos na fase de Compreensão do Negócio.O feedback dos usuários finais e das partes interessadas é coletado para avaliar a eficácia e a utilidade dos resultados.
Com base no feedback recebido, ajustes podem ser feitos nos modelos, nos processos ou nas estratégias de negócios para melhorar continuamente os resultados.

  A inclusão da fase de Feedback é fundamental para garantir que os projetos de mineração de dados continuem a agregar valor ao negócio ao longo do tempo, adaptando-se às mudanças nas necessidades e circunstâncias. O feedback contínuo permite a melhoria iterativa dos modelos e processos, garantindo que permaneçam alinhados com os objetivos e metas em evolução da organização.
  
  O CRISP-DM é um modelo iterativo, o que significa que as fases não são lineares e podem exigir revisões ou iterações à medida que o projeto avança. Ele fornece uma estrutura flexível que pode ser adaptada às necessidades específicas de cada projeto de mineração de dados.

## Funções necessárias para o projeto.


```python
def eda(df:pd.DataFrame)->None:
  print("-"*15,"DataFrame shape", "-"*15 )
  print(f"Linhas: {df.shape[0]}, Colunas: {df.shape[1]}")
  print("-"*15,"DataFrame info", "-"*15 )
  print(df.info())
  print("-"*15,"DataFrame isna", "-"*15 )
  print(df.isna().sum())
  print("-"*15,"DataFrame is duplicated", "-"*15 )
  print(df.duplicated().sum())
  print("-"*15,"DataFrame describe numeric", "-"*15 )
  print(df.describe(percentiles=[.01,0.1,.25,.5,.75,.90,.99]))
  print("-"*15,"DataFrame desribe categorical", "-"*15 )
  print(df.describe(exclude=[np.number]))

def perc_per_level(df:pd.DataFrame)->pd.DataFrame:
  df=(df.groupby("experience_level")
  .size()
  .reset_index(name="qtd_level")
  .assign(perc_per_level=lambda x:(x['qtd_level'] / x['qtd_level'].sum()) )
  .sort_values(by="perc_per_level", ascending=False))
  return df


def is_outlier(array, extreme=False):
  """função para identificar se na coluna há outliers"""
  q1,q3 = np.quantile(array, [.25,.75])
  iqr = q3-q1

  factor = 3. if extreme else 1.5
  upper_outlier = q3 + factor*iqr
  lower_outlier = q1 - factor*iqr

  return (array < lower_outlier) | (array > upper_outlier)
```

```python
qtd_senioridade_pan=periodo_pandemia['experience_level'].value_counts().to_frame().reset_index()
#anos: 2020,2021 e 2022

qtd_senioridade=periodo_sem_pandemia['experience_level'].value_counts().to_frame().reset_index()
#ano:2023
```

```python
(pd.merge(qtd_senioridade_pan,qtd_senioridade,
         how="inner", on="index")
.set_index("index")
.rename({"experience_level_x":"qtd_pand",
        "experience_level_y":"qtd_nao_pand"},axis=1))\
.style.background_gradient(cmap="YlGnBu")
```

```python
salario_pandemico=np.round(
(periodo_pandemia.groupby("job_category")
.agg(salario_medio_pan=("salary_in_usd","mean"))
.reset_index()),2)

salario_nao_pandemico=np.round((periodo_sem_pandemia.groupby("job_category")
.agg(salario_medio=("salary_in_usd","mean"))
.reset_index()),2)
```

```python
#comparativos dos salário na pandemia e depois da pandemia
'''
De acordo com a presente análise temos uma visão mais clara sobre as flutuações de salário na época
pandêmica e pós pandêmico. Com de maiores acréscimos para Machine Learning e AI, seguidos de BI e visualization
porém alguns setores sofreram com quedas tais como Cloud and database, Data Management and Strategy.
'''
salario_comparativo=pd.merge(salario_pandemico, salario_nao_pandemico, how="inner", on="job_category")
salario_comparativo
salario_comparativo["diferenca_salarial_pct"] = salario_comparativo.apply(lambda x: ((x["salario_medio"] - x["salario_medio_pan"]) / x["salario_medio"]) * 100, axis=1)
salario_comparativo.style.background_gradient(axis=0, cmap='YlGnBu_r')
```
